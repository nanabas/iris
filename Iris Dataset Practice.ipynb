what is the simplest model we can think of?

# Modelling - Simple Manual Model
Lets manually look at our data visualized earlier to decide some cutoff points for classification

````python
def petal_length_prediction(petal_length):
    # Predicts iris species given petal length
    if petal_length < 2.5:
        return 0
    elif petal_length < 4.8:
        return 1
    else:
        return 2
````

````python
df_train['petal length (cm)'].values
````

````python
x_train[:, 2] # same as petal length values in the training dataset DataFrame
````

````python
manual_y_predictions = np.array([petal_length_prediction(val) for val in x_train[:, 2]])
````

````python
manual_model_accuracy = np.mean(manual_y_predictions == y_train)
````

````python
print(f"Manual model accuracy: {manual_model_accuracy * 100:.2f}%")
````

# Modeling - Logistic Regression

````python
from sklearn.linear_model import LogisticRegression
````

### Using cross-validation to evalute our model

````python
df_predictions = df_train.copy()
````

````python
def incorrect_pred_plot(df_predictions, x_axis_feature, y_axis_feature):
    fig, axs = plt.subplots(2, 2, figsize=(10,10))
    axs = axs.flatten()
    sns.scatterplot(x=x_axis_feature, y=y_axis_feature, data=df_predictions, hue='prediction_label', ax=axs[0])
    axs[0].set_title("Predictions")
    sns.scatterplot(x=x_axis_feature, y=y_axis_feature, data=df_predictions, hue='target_name', ax=axs[1])
    axs[1].set_title("Actual Classifications")
    sns.scatterplot(x=x_axis_feature, y=y_axis_feature, data=df_predictions, hue='correct_prediction',hue_order=[True, False], ax=axs[2])
    axs[3].set_visible(False)
    plt.show()
````

````python
incorrect_pred_plot(df_predictions, 'petal length (cm)', 'petal width (cm)')
````

# Model Tuning

````python
for reg_param in (0.1, 0.5, 1, 1.5, 1.7, 1.9, 2, 2.1, 2.3):
    print(reg_param)
    model = LogisticRegression(max_iter=200, C=reg_param)
    accuracies = cross_val_score(model, x_train, y_train, cv=5, scoring='accuracy')
    print(f"Accuracy: {np.mean(accuracies) * 100:.2f}%")
````

# Final Model - LR

````python
model = LogisticRegression(max_iter=200, C=2)
````

````python
x_train = df_train.drop(columns=['target', 'target_name', 'petal length (cm)','petal width (cm)' ]).values
x_test = df_test.drop(columns=['target', 'target_name', 'petal length (cm)','petal width (cm)' ]).values
y_train = df_train['target'].values
y_test = df_test['target'].values
````

````python
print(f" Accuracy: {(np.mean(svm_pred == y_test)) * 100 :.2f}%")
````

### visualizing SVM

````python
# set max and mins for graph
x_min, x_max = x_test[:, 0].min() - 1, x_test[:,0].max() +1
y_min, y_max = x_test[:, 1].min() - 1, x_test[:,1].max() +1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02), np.arange(y_min, y_max, 0.02))
# flatten grid points so they can be passed into predict
grid_points = np.c_[xx.ravel(), yy.ravel()]
Z = clf_linear.predict(grid_points)
Z = Z.reshape(xx.shape)
````

````python
plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.coolwarm)
plt.scatter(x_test[:, 0], x_test[:, 1], c=y_test, s=30, edgecolors='k', cmap=plt.cm.coolwarm)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('SVC for Iris Flower Classification')
plt.show()
````
